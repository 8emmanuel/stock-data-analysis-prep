```python
import pandas as pd
```

```python
import numpy as np
```

```python
# Load dataset
df = pd.read_csv('2018_Financial_Data.csv')
```

```python
# Show first 5 rows
df.head()
```

```python
# Dataset info
df.info()
```

```python
# Quick stats for numeric columns
df.describe()
```

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# Drop columns where the number of non-NaN values is more than 50%
# This allows us to get rid of columns that will not contribute meaningfully to the dataset
threshold = len(df) * 0.5
df_new = df.dropna(axis=1, thresh= threshold)
```

```python
df_new.describe()
```

```python
# Inputting median values into numeric columns which do not have values
numeric_cols = df_new.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    median_value = df_new[col].median()
    df_new[col].fillna(median_value, inplace=True)
```

```python
# Inputting "missing" to replace categorical columns which do not have values
categorical_cols = df_new.select_dtypes(include=['object']).columns

df[categorical_cols] = df_new[categorical_cols].fillna("Missing")
```

```python
# Check again for missing values
print("Missing values per column:")
print(df_new.isnull().sum())
```

```python
#Save cleaned dataset for future use
df_new.to_csv('2018_Financial_Data_Cleaned.csv', index=False)
```
